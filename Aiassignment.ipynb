```python
from IPython.core.display import display, HTML
display(HTML("<style>div.output_scroll { height: 44em; }</style>"))
import networkx as nx
import matplotlib.pyplot as plt
G = nx.DiGraph()
g=[['oradea','zerind'],['oradea','sibiu'],['arad','zerind'],['arad','sibiu'],['arad','timisora'],
   ['lugoj','timisora'],['lugoj','mehadia'],['dobreta','mehadia'],['dobreta','craiova'],
   ['craiova','piteshi'],['craiova','rimnicu vilcea'],['rimnicu vilcea','sibiu'],
   ['sibiu','fagaras'],['fagaras','bucharest'],['giurgiu','bucharest'],['piteshi','bucharest'],
   ['bucharest','urziceni'],['urziceni','hirsova'],['hirsova','eforie'],['urziceni','vaslui'],
   ['iasi','vaslui'],['iasi','neamt']]
#print(g)
G.add_edges_from(g)
plt.figure(0,figsize=(13,13)) 
nx.draw(G,with_labels=True,node_size=800,node_color='b',font_size=16,arrowhead_length=1)
plt.show()
marked=[]
open=[]
explore=[]
limitincreaser=0
print("Enter Source")
source=input()     #'dobreta'
print("Enter Destination")
destination=input()      #'bucharest'

cont=True
dynamicnode=[]
dynamicgraph = nx.DiGraph()
open.append(source)
val_map={source: 'r',
        destination:'g'}
while(cont):
    dynamicgraph.clear()
    dynamicgraph.add_node(source,node_color="r")
    limitincreaser+=1
    limit=limitincreaser
    actuallimit=limitincreaser
    open.clear()
    marked.clear()
    explore.clear()
    open.append(source)
    #print("actual limit::",actuallimit)
    print("-------------------------------------Starting from limit 1 again---------------------------------")
    while(limit!=0):
        plt.figure(actuallimit,figsize=(6,6)) 
        #pos=nx.circular_layout(dynamicgraph)
        values = [val_map.get(node,'b') for node in dynamicgraph.nodes()]
        nx.draw(dynamicgraph,with_labels=True,node_size=800,node_color=values,font_size=16)
        #plt.show()
        #nx.draw_networkx_nodes(dynamicgraph,with_labels=True, nodelist=marked,node_color="b")
        plt.show()
        for j in range(len(open)):
            if(destination==open[j]):
                print("-------------------------------------found at limit",actuallimit,"---------------------------------")
                cont=False
                break
            marked.append(open[j])
            explore.append(open[j])
        open.clear()
        limit-=1
        #print(limit)
        if(limit!=0 and len(explore)!=0):
            #print("here")
            for i in range(len(explore)):
                for j in range(len(g)):
                    if(g[j][0]==explore[i]):
                        if((g[j][1] not in open) and (g[j][1] not in marked) ):
                            dynamicgraph.add_edge(g[j][0],g[j][1],color="#dd4b39")
                            open.append(g[j][1])
            #print(open)
            explore.clear()
    if(limitincreaser==30):
        print("ELEMENT NOT REACHABLE OR PRESENT")
        break
```
